```python
indA = pd.Index([1, 3, 5, 7, 9])
indB = pd.Index([2, 3, 5, 7, 11])

s1 = pd.Series([1,2,3,4,5],index=indA)
s2 = pd.Series([1,2,3,4,5],index=indB)
s1,s2
```

```python
# 交集：提取共同拥有的索引
indC = indA.intersection(indB)         # 返回 Index([3, 5, 7], dtype='int64')
s3 = pd.Series(s1+s2,index=indC)
s3
```

```python
# 1. 常规的单/多重条件数值筛选
s = pd.Series([10, 20, 30, 40, 50], index=['a', 'b', 'c', 'd', 'e'])
# 筛选出大于15 且 小于45 的值 (注意：多条件必须加括号！)
condition = (s > 15) & (s < 45)
print(condition)
```

```python
# 1. 由单个Series创建(DataFrame变成只有一列的表)
pop_s = pd.Series({'CA': 3833, 'TX': 1966})
print(pd.DataFrame(pop_s, columns=['population']))
```

```python
# 2. 由Series组成的字典创建(每个Series变成一列，自动按索引对齐)
area_s = pd.Series({'CA': 423967, 'TX': 695662})
print(pd.DataFrame({'population': pop_s, 'area': area_s}))
```

```python
# 3. 由字典构成的列表创建(每个字典变成一行，键变成列名；缺失的键填入 NaN)
data_list = [{'a': 1, 'b': 2}, {'b': 3, 'c': 4}]
print(pd.DataFrame(data_list))
# 结果中第一行的 'c' 为 NaN，第二行的 'a' 为 NaN
```

```python
# 4. 由二维 NumPy 数组创建(可人为指定 index 和 columns)
arr = np.random.rand(3, 2)
print(pd.DataFrame(arr, columns=['foo', 'bar'], index=['a', 'b', 'c']))
```
