```python
from pathlib import Path
from transformers import BertTokenizer, BertModel

# 【HuggingFace模型加载机制说明】
# 情况A：传入的是本地路径(例如：r"D:\bert-base-chinese")
# 当你传入明确的本地路径时，代码会认为：“模型已经在这个文件夹里了，我直接去读就行”。
# 如果这个文件夹不存在，或者里面没有 config.json 等关键文件，它就会引发 OSError，
# 它绝对不会帮你自动下载到这个你指定的本地路径下。
# 情况B：传入的是模型ID/仓库名(例如："bert-base-chinese")
# 代码发现这不像一个本地路径，就会去HuggingFace官网寻找同名仓库并自动下载。
# 注：默认下载的位置是系统默认的隐藏缓存文件夹(Windows是C:\Users\用户名\.cache\huggingface\hub)。
# 解决方案：结合情况B，传入模型ID，并使用 cache_dir 参数接管默认下载位置。

# 模型存储文件夹
cache_path = Path(r"D:\HuggingfaceModels")

# 1. 传入模型在官网的ID(名字)
# 2. 使用cache_dir参数指定下载并缓存到指定文件夹
tokenizer = BertTokenizer.from_pretrained(
    "bert-base-chinese",
    cache_dir=cache_path, # cache_dir:指定下载位置
    trust_remote_code=True# 确定模型来源可靠
)

model = BertModel.from_pretrained(
    "bert-base-chinese",
    cache_dir=cache_path, # cache_dir:指定下载位置
    trust_remote_code=True# 确定模型来源可靠
)

text = "你是谁？"
# 3. 将text输入分词和编码模型
encoded_input = tokenizer(text, return_tensors='pt')
# return_tensors:把分词后的结果打包成哪种深度学习框架专属的数据格式
# pt：代表 PyTorch
# tf：代表 TensorFlow
# np：代表 NumPy
# 如果不加这个参数，Tokenizer默认返回的是Python的列表(List)
print(encoded_input)
# {'input_ids': tensor([[ 101,  872, 3221, 6443, 8043,  102]]),
#  'token_type_ids': tensor([[0, 0, 0, 0, 0, 0]]),
#  'attention_mask': tensor([[1, 1, 1, 1, 1, 1]])}
# (1).input_ids(输入 ID)：每个字的索引(字典里对应的数字)。101和102是雷打不动的开始和结束标志。
# (2).token_type_ids(句子类型ID)：第几句话，编号从0开始。
# (3).attention_mask(注意力掩码)：1代表这是真实的字，请模型注意它；0代表这是凑数的，请模型忽略它。

# 4. 将编码好的文字输入给预训练好的bert
output = model(**encoded_input) # **是解包的意思
print(output)
```

```python
import torch
import torch.nn as nn

# 假设词汇表大小（包括特殊标记如SOS和EOS）
vocab_size = 10
embedding_dim = 4

# 创建嵌入层
embedding_layer = nn.Embedding(vocab_size, embedding_dim)

# 假设索引0是SOS，索引1是EOS
SOS_token = 0
EOS_token = 1

# 目标序列的索引表示
target_sequence = [2, 3, 4, 5, 6]  # 假设 "it", "was", "the", "best", "of"

# 添加起始标记和结束标记
decoder_input = [SOS_token] + target_sequence
decoder_output = target_sequence + [EOS_token]

# 转换为张量
decoder_input_tensor = torch.tensor(decoder_input)
decoder_output_tensor = torch.tensor(decoder_output)

# 嵌入
embedded_decoder_input = embedding_layer(decoder_input_tensor)
embedded_decoder_output = embedding_layer(decoder_output_tensor)

print("Decoder Input (with SOS):", decoder_input_tensor)
print("Decoder Output (with EOS):", decoder_output_tensor)
print("Embedded Decoder Input:", embedded_decoder_input)
print("Embedded Decoder Output:", embedded_decoder_output)
```
